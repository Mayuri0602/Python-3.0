```python
# Matplotlib is basic and powerful plotting library in Python 

import matplotlib.pyplot as plt
```

```python
# line plot
import pandas as pd
line_data = pd.DataFrame({
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "June", "July", "Aug", "Sep", "Oct", "Nov", "Dec"],
    "Revenue": [5000, 7000, 8000, 7500, 9000, 10000, 9500, 11000, 10500, 11500, 12000, 12500]
})

plt.plot(line_data['Month'], line_data['Revenue'])
plt.grid()
plt.title("Month vs Revenue Graph")
plt.xlabel("Months")
plt.ylabel("Revenue")
```

```python
# BarChart
# categories vs numerical values
import pandas as pd

bar_data = pd.DataFrame({
    "Product": ["A", "B", "C", "D", "E"],
    "Sales": [120, 340, 230, 150, 200]
})

plt.bar(bar_data['Product'], bar_data['Sales'], color = 'Blue', edgecolor = 'Black')
# plt.grid()
plt.title("Product and Sales Graph")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()
```

```python
# ScatterPlot
# showing the relationship between two numerical variables 

import numpy as np

scatter_data = pd.DataFrame ({
    "Height_cm": [145, 125, 130, 155, 147, 153],
    "Weight_kg": [50, 45, 60, 65, 64, 68]
})

plt.scatter(scatter_data["Height_cm"], scatter_data["Weight_kg"])
plt.title("Height_cm and Weight_kg Graph")
plt.xlabel("Height_cm")
plt.ylabel("Weight_kg")
plt.show()
```

```python
# PieChart
# Showing the parts of a whole (Proportions)

pie_data = pd.DataFrame({
    "Company": ["Apple", "Samsung", "Xiaomi", "OnePlus", "Others"],
    "MarketShare": [30, 25, 20, 15, 10]
}) 

plt.pie(pie_data['MarketShare'], labels = pie_data['Company'], autopct = '%1.1f%%')
plt.title("MobilePhone Company Distribution over MarketShare")
plt.show()
```
