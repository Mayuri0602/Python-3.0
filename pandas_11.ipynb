```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
# getting the dataset ehich are into the seaborn
sns.get_dataset_names()
```

```python
# loading the datasets
df = sns.load_dataset('tips')
df
```

```python
# Scatter plot in seaborn on total bill and tip
plt.figure(figsize=(5, 3))
sns.scatterplot(x=df['total_bill'], y=df['tip'], data=df)
plt.title("Scatter plot between the total bill and the tip")
plt.xlabel("Total bill")
plt.ylabel("Tip")
# plt.grid()
plt.show()
```

```python
# Line plot used for the trends or continuous data(size, tip)
plt.figure(figsize=(5,3) )
sns.lineplot(x=df['size'], y=df['tip'], data=df, marker='o')     # 'o' --> '*'
plt.title("Line plot of size vs tip")
plt.xlabel('Size')
plt.ylabel('Tip')
plt.show()
```

```python
# barplot
sns.barplot(x=df['day'], y=df['total_bill'], data=df)
plt.title("Plotting the bar plot between the days and the total bill")

# black lines represent error bars thst shows uncertainity in data 
```

```python
# Countplot - counts frequency of categories (like histogram but for categories)
plt.figure(figsize=(5,3))
sns.countplot(x=df['smoker'], data=df)
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.countplot(x=df['time'], data=df)
plt.title("Countplot of time")
plt.xlabel('Time')
plt.ylabel('frequency')
plt.show()

# At the time of dinner frequency is larger than the daytime
```

```python
plt.figure(figsize=(5,3))
sns.countplot(x=df['day'], data=df)
plt.title("Countplot of day")
plt.xlabel('Day')
plt.ylabel('frequency')
plt.show()

# We analyse that count of sat is largest among all and the least is on fri,
# on sun, frequency is larger than thurs but lesser than sat
```

```python
plt.figure(figsize=(5,3))
sns.countplot(x=df['sex'], data=df)
plt.title("Countplot of sex")
plt.xlabel('Sex')
plt.ylabel('frequency')
plt.show()

# From this, we analyse that the count of men is greater than the female
```

```python
sns.boxplot(x=df['total_bill'], data=df)
plt.show()
```

```python
sns.boxplot(x=df['tip'], data=df)
plt.title('Boxplot of tip')
plt.show()
```

```python
plt.figure(figsize=(10, 20))
arg, axs  = plt.subplots(2,2)
axs[0,0].boxplot(df['total_bill'])
axs[0,1].boxplot(df['size'])
axs[1,0].scatter(df['total_bill'], df['size'])
axs[1,1].scatter(df['total_bill'], df['tip'])
```
