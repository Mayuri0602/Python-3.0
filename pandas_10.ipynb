```python
# BoxPlot :- 
import seaborn as sns
import matplotlib.pyplot as plt

boxplot_data = pd.DataFrame({
    "Class": ["A"]*10 + ["B"]*10 + ["C"]*10,
    "Score": [56, 67, 45, 78, 89, 90, 76, 65, 59, 83,
              72, 85, 69, 95, 88, 91, 73, 68, 80, 77,
              60, 70, 75, 82, 79, 88, 93, 84, 77, 85]
})
# plt.figure(figsize=(8, 6))
plt.boxplot([boxplot_data[boxplot_data['Class']==c]['Score'] for c in ['A', 'B']], 
           labels=['A', 'B'])
plt.title("Distribution of Scores by Class")
plt.xlabel("Class")
plt.ylabel("Score")
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

sns.boxplot(x=boxplot_data['Class'], y=boxplot_data['Score'], data=boxplot_data)
```

```python
x = [1,2,3,4,5]
y = [1,4,6,8,10]

plt.fill_between(x, y, color="skyblue", alpha=0.4)
plt.plot(x, y, color="blue")
plt.title("Area Plot")
plt.show()
```

```python
fig, axs = plt.subplots(2, 2)


# Plotting the line plot
axs[0,0].plot([1,2,3],[1,4,9])
axs[0,0].set_title("Line")


axs[0,1].bar(['A','B','C'], [3,5,7])
axs[0,1].set_title("Bar")

axs[1,0].scatter([1,2,3],[5,7,9])
axs[1,0].set_title("Scatter")

axs[1,1].hist([1,2,2,3,3,4,4,5,6], bins=5)
axs[1,1].set_title("Histogram")

plt.tight_layout()
plt.show()
```
