```python
import pandas as pd
```

```python
df = pd.read_csv("titanic - titanic.csv")
df
```

```python
# 1. Load the Titanic dataset into a Pandas DataFrame. Check the first 5 rows.
df.head()
```

```python
# 2. Find the shape (rows & columns) of the dataset.
# 3. Get column names and their data types.
df.shape
print(df.shape)

df.dtypes
print(df.dtypes)
```

```python
# 4. Check for missing values in each column.
# 5. Drop the column Ticket (since it is not useful for analysis).
print(df.isna().sum())         # isnull()

df = df.drop(columns=['Ticket'], errors='ignore')
print(df.dtypes)
```

```python
# 6. Fill missing values in Age with the mean age.
# 7. Fill missing values in Embarked with the most frequent value.
df['Age'] = df['Age'].fillna(df['Age'].mean())
print(df['Age'])
# print(df.isna().sum())

df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])      # [0] :- actual value (not a Series)
print(df['Embarked'])
```

```python
# 8. Create a new column FamilySize = SibSp + Parch + 1.
# 9. Drop the column Cabin (too many missing values).
# 10. Rename the column Pclass to PassengerClass.
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
print(df['FamilySize'])
print(df.dtypes)

df = df.drop(columns=['Cabin'], errors='ignore')
print(df.dtypes)

df = df.rename(columns = {'Pclass' : 'PassengerClass'})
print(df.dtypes)
```

```python
# 1. Select all passengers who were in PassengerClass 1.
P_class1 = df[df['PassengerClass'] == 1]
print(P_class1)

# 2. Select all passengers who are older than 50 years.
older = df[df['Age'] > 50]
print(older)

# 3. Find all female passengers who survived.
female_survived = df[(df['Sex'] == 'female') & (df['Survived'] == 1)]
print(female_survived)

# 4. Find all passengers who paid a fare greater than 50.
fare50 = df[df['Fare'] > 50]
print(fare50)

# 5. Find the passengers who were traveling alone (FamilySize == 1).
p = df[df['FamilySize'] == 1]
print(p)
```

```python
# 1. Find the average age of passengers grouped by Sex.
avg_pass_sex = df.groupby('Sex')['Age'].mean()
avg_pass_sex

# 2. Find the survival rate grouped by PassengerClass.
s = df.groupby('PassengerClass')['Survived'].mean()      # survival rate = fraction of passengers who survived, sum of 1s/Total count
s

# 3. Find the maximum fare paid by each PassengerClass.
m = df.groupby('PassengerClass')['Fare'].max()
m

# 4. Count how many passengers boarded from each port (Embarked).
c = df['Embarked'].value_counts()
c

# 5. Find the average family size for survivors vs non-survivors.
a = df.groupby('Survived')['FamilySize'].mean()
a
```

```python
# 1. Sort the dataset by Fare in descending order.
s = df.sort_values(by='Fare', ascending=False)
s

# 2. Find the top 5 oldest passengers.
s = df.sort_values(by='Age', ascending=False)
oldest = s.head()
print(oldest)

# 3. Find the top 5 passengers who paid the highest fare.
t = df.sort_values(by= 'Fare', ascending=False)            
highest_fare = t.head()                           # highest_fare = df.nlargest(5, 'Fare')
print(highest_fare)

# 4. Find the survival rate of children (Age < 18) vs adults.
df['child'] = df['Age'] < 18
s = df.groupby('child')['Survived'].mean()
s
```

```python
# 5. Create a new column FarePerPerson = Fare/FamilySize and find the avg FarePerPerson for each PassengerClass.
df['FarePerPerson'] = df['Fare'] / df['FamilySize']

avg = df.groupby('PassengerClass')['FarePerPerson'].mean()
avg
```
